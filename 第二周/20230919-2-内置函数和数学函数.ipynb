**<font color = blue> 内置函数：print(表达式, ......) </font>**  
> + (逐个) 求值实参表达式，并在执行窗口 (shell) 中输出  
> +  默认情况下，每两项输出之间用**空格**分隔，所有输出结束后换一行
> + print 函数的更多使用细节之后介绍，也可自行查阅语言手册

```python
print(123)
```

```python
print(123, 1.23)
```

```python
print(1/17) ## 调用 print 时，实参可以是任意合适的表达式

## Python 程序里的一行里，从 # 开始到该行末尾都是注释
## 注释是说明性文字，供人阅读
## 执行时，解释器将跳过所有注释 (没有语义)
```

**<font color=blue>print 输出浮点数时，可指定保留小数点后有效数字的位数 (即格式化输出)</font>**  
> + 下面只介绍一种简单处理方式 (有点过时)；
> +  其它格式化方式 (例如 f-string, str.format 方法) 之后逐步介绍，也可自行查阅语言手册

```python
print('%.6f' % (1/17)) ## 保留小数点后 6 位有效数字
```

```python
print('%f'%0.5)    

## %f，称为模板字符串，表示输出一个浮点数，并且保留小数点后 6 位有效数字 
## (沿袭了 C 语言中的规则)
```

```python
print('%.1f' % 1.15) ## Note：非精确的四舍五入规则
print('%.1f' % 1.85)
```

**<font color=blue>输出整数时，也有相应的格式化方式</font>**  
> + 下面只简单介绍；详细情况之后介绍，也可自行查阅语言手册

```python
## %d，输出一个整数
print('%d' % 123)

## %10d，输出整数可指定输出宽度；整数不够宽时，默认居右输出，高位补空格
print('%10d' % 456) 
```

```python
## 输出多个数值对象
print('%d'%123, '%.6f'%.123)

print('%d %.6f' % (123, .123))

## 上述两种写法等价
```

**<font color=blue> 内置函数：abs(x) </font>**  
> + Return the absolute value of a number. 
> + The argument may be an integer or a floating point number.
> + If the argument is a complex number, its magnitude is returned. 

```python
abs(-389)
```

```python
abs(-2.3**5)
```

```python
abs((2+3.5j)**4)
```

**<font color=blue> 内置函数：max(arg1, arg2, *args[, key]) </font>** 
> + Return the largest of two or more arguments.  

**<font color=blue> 内置函数：min(arg1, arg2, *args[, key])  </font>** 
> + Return the smallest of two or more arguments.


```python
max(2, 4.3, -1.23, 8.6)
```

**<font color=blue>内置函数：pow(base, exp[, mod]) </font>**  
> + Return base to the power exp; if mod is present, return base to the power exp, modulo mod (computed more efficiently than pow(base, exp) % mod). 
> + The two-argument form pow(base, exp) is equivalent to using the power operator: base**exp.


```python
pow(2.3, 1.5)
```

```python
pow(3, 3, 2)
```

```python
pow(1.5, 2.78, 2.4)
```
